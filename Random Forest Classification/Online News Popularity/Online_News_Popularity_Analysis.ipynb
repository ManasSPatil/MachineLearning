#Data Preprocessing 
Here we manipulate the data so that we can apply the machine learning model

```
dataset.info()
```

```
dataset
```

Dropping uncessary columns like url and is_weekend.Url does not contribute to the popularity of the news article , and since we already have is_saturday , is_sunday columns we dont need is_weekend

```
dataset=dataset.drop(columns=['url','is_weekend'],axis=1)
dataset
```

To make a classifier we make to change the values in the share column.

Shares>1400 -->1

Shares<=1400 -->0

```
dataset.loc[dataset.shares<=1400,'shares']=0
dataset.loc[dataset.shares>1400,'shares']=1
```

```
dataset
```

Extracting the matrix of features and the dependent variable

```
X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

Spltting the dataset into training and test set. We don't need encoding and dummy coding since it is already done and all the vairables are integers or floats. 

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

Feature scaling matrix of features so that the algorithm is less compute intensive and the values are standardised 

```
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
X_train=sc_X.fit_transform(X_train)
X_test=sc_X.transform(X_test)
```

Now applying PCA on the dataset to see the contributions of each column, we select the components which explain the most variance.

```
from sklearn.decomposition import PCA
pca=PCA(n_components=31)
X_train=pca.fit_transform(X_train)
X_test=pca.transform(X_test)
explained_variance=pca.explained_variance_ratio_
```

```
X_train
```

#Applying Random Forest Classifier

```
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=500,criterion='entropy', random_state=1)
classifier.fit(X_train,y_train)
```

#Predicting values and checking accuracy

Making predictions

```
y_pred=classifier.predict(X_test)
```

Applying K-fold cross validation



```
from sklearn.model_selection import cross_val_score
accuracies=cross_val_score(estimator=classifier,X=X_train,y=y_train,cv=10)
accuracies.mean()
```

```
accuracies.std()
```

Making a confusion matrix

```
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```
